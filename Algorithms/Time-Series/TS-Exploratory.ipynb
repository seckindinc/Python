```python
#importing libraries
import pandas as pd
import pandas_profiling
from pandas.plotting import lag_plot
import matplotlib.pyplot as plt
from pandas.plotting import autocorrelation_plot
```

```python
#reading data set and converting string date column into datetime
df = pd.read_csv('../../Data-Sets/female_birth_ts.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date',inplace=True)
```

```python
#df report
df.profile_report()
```

## __Line Plot__

```python
df.plot()
```

```python
#resampling to see weekly changes
weekly_df = df.resample('W').apply(sum)
weekly_df.plot()
```

```python
#resampling too see monthly changes
monthly_df = df.resample('M').apply(sum)
monthly_df.plot()
```

## __Histogram__

```python
df.hist()
```

```python
df.plot(kind='kde')
```

## __Box Plots__

```python
df.boxplot()
```

```python
df.reset_index(inplace=True)
df['month'] = pd.DatetimeIndex(df['Date']).month
df.pivot(index='Date', columns='month', values='Births').boxplot(figsize = (20, 10))
```

## __Lag Plot__

```python
for i in range(1,10):
    lag_plot(df.Births, lag=i)
    plt.figure()
```

## __Autocorellation Plot__

```python
x = autocorrelation_plot(df.Births)
```
