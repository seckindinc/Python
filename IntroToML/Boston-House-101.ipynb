__Column Definitions__

1. CRIM: per capita crime rate by town 
2. ZN: proportion of residential land zoned for lots over 25,000 sq.ft. 
3. INDUS: proportion of non-retail business acres per town 
4. CHAS: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise) 
5. NOX: nitric oxides concentration 
6. RM: average number of rooms per dwelling 
7. AGE: proportion of owner-occupied units built prior to 1940 
8. DIS: weighted distances to ﬁve Boston employment centers 
9. RAD: index of accessibility to radial highways 
10. TAX: full-value property-tax rate per 10000 
11. PTRATIO: pupil-teacher ratio by town 
12. B: 1000(Bk−0.63)2 where Bk is the proportion of blacks by town 
13. LSTAT: % lower status of the population 
14. MEDV: Median value of owner-occupied homes in $1000s

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
X = data_set[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']]
y = data_set['MEDV']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
def pretty_print_linear(coefs, names = None, sort = False):
    if names == None:
        names = ["X%s" % x for x in range(len(coefs))]
    lst = zip(coefs, names)
    if sort:
        lst = sorted(lst,  key = lambda x:-np.abs(x[0]))
    return " + ".join("%s * %s" % (round(coef, 3), name)
                                   for coef, name in lst)
```

```python
linear_regression = LinearRegression()
model = linear_regression.fit(X_train_scaled, y_train)
print('The accuracy of the Linear Regression is: {:.2f}'.format(model.score(X_train_scaled,y_train)))
print('The accuracy of the Linear Regression is: {:.2f}'.format(model.score(X_test_scaled,y_test)))
pretty_print_linear(model.coef_)
```

```python
ridge = Ridge()
model = ridge.fit(X_train_scaled, y_train)
print('The accuracy of the Ridge Regression is: {:.2f}'.format(model.score(X_train_scaled,y_train)))
print('The accuracy of the Ridge Regression is: {:.2f}'.format(model.score(X_test_scaled,y_test)))
pretty_print_linear(model.coef_)
```

```python
lasso = Lasso()
model = lasso.fit(X_train_scaled, y_train)
print('The accuracy of the Lasso Regression is: {:.2f}'.format(model.score(X_train_scaled,y_train)))
print('The accuracy of the Lasso Regression is: {:.2f}'.format(model.score(X_test_scaled,y_test)))
pretty_print_linear(model.coef_)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor().fit(X_train_scaled, y_train)
print('The accuracy of the Decision Tree Regression is: {:.2f}'.format(model.score(X_train_scaled,y_train)))
print('The accuracy of the Decision Tree Regression is: {:.2f}'.format(model.score(X_test_scaled,y_test)))
```
