```python
#Pandas get_dummies automatically creates one-hot encodings for all categorical variables
data_dummies = pd.get_dummies(data)
```

```python
data_dummies.head()
```

```python
pd.get_dummies(data['education.num']).head()
```

```python
#Slicing columns to be used in classification except income column
features = data_dummies.loc[:, 'age':'native.country_Yugoslavia']
```

```python
#Converting pandas to numpy array
X = features.values
y = data_dummies['income_>50K'].values
print('X shape:',X.shape, ' and y shape:', y.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
print(logreg.score(X_test, y_test))
```
