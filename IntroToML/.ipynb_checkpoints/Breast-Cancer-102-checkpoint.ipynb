1. id: ID number

3. adius_mean: mean of distances from center to points on the perimeter
4. texture_mean: standard deviation of gray-scale values
5. perimeter_mean: mean size of the core tumor
6. area_mean
7. smoothness_mean: mean of local variation in radius lengths
8. compactness_mean: mean of perimeter^2 / area - 1.0

11. symmetry_mean
12. fractal_dimension_mean: mean for "coastline approximation" - 1

```python
data_set = pd.read_csv('../Data-Sets/breast_cancer.csv')
```

```python
data_set.head()
```

```python
data_set.describe()
```

```python
y = data_set['diagnosis']
del data_set['diagnosis']
del data_set['Unnamed: 32']
del data_set['id']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(data_set)
data_set_scaled = pd.DataFrame(scaler.transform(data_set), columns=data_set.columns)
```

```python
data_set_scaled.head()
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 2, min_samples = 10)
clusters = dbscan.fit_predict(data_set_scaled)
```

```python
from sklearn.metrics.cluster import silhouette_score
```

```python
silhouette_score(data_set_scaled,clusters)
```

```python
for i in [0.1, 0.5, 1, 3, 5, 7, 9, 11, 13]:
    dbscan = DBSCAN(eps = i, min_samples = 10)
    clusters = dbscan.fit_predict(data_set_scaled)
    print('Number of clusters: ',len(np.unique(clusters)))
    print('Cluster sizes: ',np.bincount(clusters+1))
    print('')
```
